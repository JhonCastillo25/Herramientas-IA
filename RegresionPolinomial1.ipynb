```
import numpy as np
import matplotlib.pyplot as plt
N=50
x=6*np.random.rand(N,1)-3
y=0.5*x**2+x+2+0.5*np.random.randn(N,1)
plt.plot(x,y,"ro")
plt.show()
```

```
Phi=np.c_[np.ones((N,1)),x,x*x]
print(Phi)
#Forma Opcional
#Phi3=np.c_[np.ones((N,1)),x,x**2]
#print(Phi3)
```

```
Theta=np.linalg.inv(Phi.T.dot(Phi)).dot(Phi.T).dot(y)
print(Theta)
```

```
xn=np.linspace(-3,3,25)
xnew=xn.reshape(-1,1)
```

```
Phi_new=np.c_[np.ones((25,1)),xnew,xnew*xnew]
print(Phi_new)
```

```
ypred=Phi_new.dot(Theta)
print(ypred)
```

```
plt.plot(xnew,ypred,"b-")
plt.plot(x,y,"ro")
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
N=50
x2n=np.linspace(-3,3,N)
x2=x2n.reshape(-1,1)
y2=0.5*x2**2+x2+2+np.random.randn(N,1)
plt.plot(x2,y2,"ro")
plt.show()
```

```
from sklearn.preprocessing import PolynomialFeatures
poly_features=PolynomialFeatures(degree=2,include_bias=False)
X_poly=poly_features.fit_transform(x2)
print(X_poly)
```

```
Lin_Reg=LinearRegression()
Lin_Reg.fit(X_poly,y2)
print(Lin_Reg.intercept_)
print(Lin_Reg.coef_)
```

```
N2=100
x2n=np.linspace(-3,3,N2)
xnew2=x2n.reshape(-1,1)
X_poly_p=poly_features.fit_transform(xnew2)
```

```
ypred2=Lin_Reg.predict(X_poly_p)
#print(ypred2)
```

```
plt.plot(xnew2,ypred2,"b-")
plt.plot(x2,y2,"ro")
plt.show()
```
