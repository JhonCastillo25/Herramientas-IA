```
import numpy as np
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import plotly.graph_objects as go

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)
print(y.shape)
print(Y.shape)
R = (5+3*X + 4*Y)+5*np.random.randn(10,10)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, R, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```
Xg=X.reshape(1,-1)
print(Xg.shape)
Yg=Y.reshape(1,-1)
print(Yg.shape)
Zg=R.reshape(1,-1)
print(Zg.shape)

print(Yg[0].shape)

fig = go.Figure()
fig.add_trace(go.Scatter3d(y=Yg[0], x=Xg[0], z=Zg[0], mode='markers', marker = dict(size = 4, color='rgb(0,0,200)')))
fig.update_layout(title = "Datos",
                  width=500, height=500, scene = dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```

```
Xphi=X.reshape(-1,1)
Yphi=Y.reshape(-1,1)
Z=R.reshape(-1,1)
Phi=np.c_[np.ones((100,1)),Xphi,Yphi]
print(Phi.shape)
```

```
Theta_Mat=np.linalg.inv(Phi.T.dot(Phi)).dot(Phi.T).dot(Z)
print(Theta_Mat)
```

```
Zpred=Phi.dot(Theta_Mat)
print(Zpred.shape)
Rpred=Zpred.reshape(10,10)
figure = plt.figure(1, figsize = (12, 4))
subplot3d2 = plt.axes(projection='3d')

subplot3d2.plot_surface(X, Y, Rpred, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
subplot3d2.scatter3D(Xg, Yg, Zg);
plt.show()

```
